```python
#loads a CSV file, adds headers, adds a couple of columns to the dataset, then does a groupby (like a pivot table)

import pandas as pd
import numpy as np

data_folder = "C:\\Users\\elidmoore\\Documents\Personal Documents\\Data Science 2021 Course\\001 - Python Programming\\"
file_to_open = data_folder + "citytemp.CSV"
df = open(file_to_open)
data = pd.read_csv(df,names = ['City','Temp','Unit'])

data['F_Temp'] = np.where(data['Unit'] == 'F', data['Temp'], (data['Temp']*(9/5)+32))
data['C_Temp'] = np.where(data['Unit'] == 'C', data['Temp'], (data['Temp']-32)/(9/5))
data['Local_Temp'] = np.round(np.where(data['Unit'] == 'F', data['F_Temp'], data['C_Temp']),2)

print(data)
data.groupby('City')['F_Temp'].agg(['sum', 'mean']).rename(columns={'sum': 'average'})
```
