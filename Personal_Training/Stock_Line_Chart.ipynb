```python
## Basic libs
import yfinance as yf
import pandas as pd
import numpy as np
import warnings

## Building Model
from sklearn import linear_model
from scipy import stats
import statsmodels
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.stats.api as sms
from statsmodels.compat import lzip

## Data Visualization
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
from sklearn import metrics


#####################################################################################

StockTicker = 'KO'
BeginningDate = "2021-01-01"
EndDate = "2021-02-05"

#initalize stock and allow you to get data from it
stock = yf.Ticker(StockTicker)
#this shows all current relevant data on stock
stock.info

#pull historical data over time
data  =  stock.history(start=BeginningDate,  end=EndDate)
#show first 5 rows of data
data.head()


#recommendations will tell you what people recommend
stock.recommendations

#pull calendar
stock.calendar

####################################################################################
#initalize data for charting using plotly
data = data.reset_index()
for i in ['Open', 'High', 'Close', 'Low']: 
      data[i]  =  data[i].astype('float64')     

#Set charting method (see all with pio.renderers)
import plotly.io as pio
pio.renderers.default = "browser"

#Candlestick Chart
from plotly.offline import plot, iplot

import plotly.graph_objects as go
fig = go.Figure(data=[go.Candlestick(
    x=data['Date'],
    open=data['Open'],
    high=data['High'],
    low=data['Low'],
    close=data['Close'])])
fig.show()

#Line Chart
#import plotly.express as px
#chartTitle = StockTicker + ' Stock Prices'
#fig = px.line(data, x="Date", y="Open", title=chartTitle)
#fig.show()
```
